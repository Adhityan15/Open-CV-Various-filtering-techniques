```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("input.jpeg", 0)

if img is None:
    print("ERROR: Image not loaded")
    exit()

# Resize for speed
img = cv2.resize(img, (256, 256))

# Apply filters
mean = cv2.blur(img, (3, 3))
gaussian = cv2.GaussianBlur(img, (3, 3), 0)
median = cv2.medianBlur(img, 3)
bilateral = cv2.bilateralFilter(img, 3, 30, 30)
laplacian = cv2.Laplacian(img, cv2.CV_64F)

# Plot results
titles = ["Original", "Mean", "Gaussian", "Median", "Bilateral", "Laplacian"]
images = [img, mean, gaussian, median, bilateral, laplacian]

plt.figure(figsize=(10, 6))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
